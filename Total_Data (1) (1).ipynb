```python
import yfinance as yf
import datetime as dt
import pandas as pd
import datetime as dt
import yfinance as yf

# Define your variables
end = dt.datetime.now()
start = end - dt.timedelta(days=730)  # Adjust start date to be within the last 730 days
company = input('STOCK : ')

# Fetch data from Yahoo Finance
df = yf.Ticker(company)
yfinance_data = df.history(start=start, end=end, interval='1h')

# Check if DataFrame is empty
if yfinance_data.empty:
    print("No data available for the specified stock or period.")
else:
    # Reset index and rename index to 'Datetime'
    yfinance_data.reset_index(inplace=True)
    yfinance_data.rename(columns={'index': 'Datetime'}, inplace=True)

    # Display the DataFrame
    print(yfinance_data.head())
```
